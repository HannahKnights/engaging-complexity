**1. (2 points) As blood moves through a vein or an artery, its velocity v is greatest along the central axis and decreases as the distance r from the central axis increases. The formula that gives v as a function of r is called the law of laminar flow. For an artery with radius 0.5 cm, the relationship between v (in cm/s) and r (in cm) is given by the function**

$$v(r) = 18500(0.25 − r^2), 0 ≤ r ≤ 0.5.$$

```python
def find_the_velocity(radius):
    assert radius >= 0 and radius <= 0.5, "radius value is out of range (0 ≤ r ≤ 0.5)"
    return 18500 * (0.25 - (radius ** 2))

#print("v(0.1) = %s" % find_the_velocity(0.1))
#print("v(0.4) = %s" % find_the_velocity(0.4))
```

**3. (2 points) A sequence is defined recursively by the given formulas. Find the first five terms of each sequence.**

**(a)** $$a_{n} =2(a_{n-1} +3)$$ $$a_{1} =4$$

```python
# Attempt 1
seq_a = [-1, 4] # we are given a1 (4), which allows us to figure out a0 (-1)

def funct(n):
    result = 2 * (seq_a[n-1] + 3)
    seq_a.append(result)

for i in range(2,6):
    funct(i)

# print('n | a')
# print('-----')
# for n, a in enumerate(seq_a):
#     print('%s | %s' % (n, a))

# Attempt 2 - after class 19/11 (same result)
# this seems a better way of doing it pythonically :)

def recur_func(n):
    if n == 1:
        return 4
    else:
        return 2 * (recur_func(n-1) + 3)

print('n | a')
print('-----')
for i in range(1,6):
    print('%s | %s' % (i, recur_func(i)))
```

**(c)** $$a_{n} = 4a_{n−1}(1 − a_{n−1})$$ $$a1 = 0.2$$

```python
# I am going to use a similar format for the recur_func function as used above, but will build the object of values `a_values`
# as I go, this is to speed up the process, which was very slow for 100 values previously, without assigning the values as we 
# iterated through the
a_values_map = {
    1 : 0.2
}
def recur_func(n):
    if n == 1:
        return 0.2
    else:
        if (n-1) in a_values_map:
            n1_value = a_values_map.get(n-1)
        else:
            n1_value = recur_func(n-1)
            a_values_map[n-1] = n1_value
        return (4 * n1_value) * (1 - n1_value)
    

# we run the function 101 times, because the function adds the n-1 value to the map, so this will ensure we have the 100th value
for i in range(1,102):
    recur_func(i)
a_values = [a_values_map[k] for k in sorted(a_values_map)]

# create the plot
n = np.linspace(1,100,100)
plt.plot(n,a_values)
plt.xlabel('n')
plt.ylabel('a')
plt.title('First 100 terms (c)')
plt.grid(True)
plt.show()
```

**5. (2 points) Plot the first 50 terms of the sequence (i) with initial condition (ii). In a different figure, plot the ratio (iii), comment on the long-time behaviour (as n becomes large) of the plots.** 
$$(i) a_{n} = a_{n−1} +a_{n−2}$$
$$(ii) a_{1} = 0, a_{2} = 1$$
$$(iii)  \frac{a_{n}}{a_{n - 1}}\ $$ 

```python
# I'll seed the ratio with a1 = 1, a2=2
a_values_map = {
    1:1,
    2:1    
}
def recur_func(n):
    if n in a_values_map:
        return a_values_map.get(n)
    else:
        if (n-1) in a_values_map:
            n1_value = a_values_map.get(n-1)
        else:
            n1_value = recur_func(n-1)
            a_values_map[n-1] = n1_value
        if (n-2) in a_values_map:
            n2_value = a_values_map.get(n-2)
        else:
            n2_value = recur_func(n-2)
            a_values_map[n-2] = n2_value
        return n1_value + n2_value
    
for i in range(1,52):
    recur_func(i)
a_values = [a_values_map[k] for k in sorted(a_values_map)]

#create the plot
n = np.linspace(1,50,50)
plt.plot(n , a_values)
plt.xlabel('n')
plt.ylabel('a')
plt.title('First 50 terms 5.(i)')
plt.grid(True)
plt.show()
```

```python
# I'll start the recurrance sequence with a1=1 and a2=1 like above
# I'll also use the same recurrant function as above - after which I'll generate ratios..
a_values_map = {
    1:1,
    2:1    
}
def recur_func(n):
    if n in a_values_map:
        return a_values_map.get(n)
    else:
        if (n-1) in a_values_map:
            n1_value = a_values_map.get(n-1)
        else:
            n1_value = recur_func(n-1)
            a_values_map[n-1] = n1_value
        if (n-2) in a_values_map:
            n2_value = a_values_map.get(n-2)
        else:
            n2_value = recur_func(n-2)
            a_values_map[n-2] = n2_value
        return n1_value + n2_value
    
for i in range(1,52):
    recur_func(i)
a_values = [a_values_map[k] for k in sorted(a_values_map)]

# the a_values are a sequence of numbers, let's use these to create the ratios
# I'll plot the ratio values on the y axis, so will call them y_values 
y_values = [1]
for i in range(2,51):
    y_value = a_values_map.get(i) / a_values_map.get(i-1)
    y_values.append(y_value)

#create the plot
n = np.linspace(1,50,50)
plt.plot(n , y_values)
plt.xlabel('x')
plt.ylabel('y')
plt.title('First 50 terms 5.(iii)')
plt.grid(True)
plt.show()
```

The difference I observe on the 2 plots is that in the first one (i) as n becomes larger the output grows exponentially, without limit, whereas on the second plot (b) as n becomes larger we see an initial 'spike' and then we see a 'levelling out' where the output becomes steady and consistent - a convergence.
